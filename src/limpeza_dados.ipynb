# Limpeza dos dados

```python
#seleção das colunas que vão ser utilizadas
colunas_utilizadas = [
    'age',
    'gender',
    'daily_social_media_time',
    'social_platform_preference',
    'stress_level',
    'sleep_hours',
    'days_feeling_burnout_per_month',

]

df_limpo = df[colunas_utilizadas].copy()
df_limpo.head()
```

Coluna 'age'

```python
#contagem de nulos na coluna 'age'
print(f"Número de valores nulos: {df_limpo['age'].isnull().sum()}")
```

```python
#remove os nulos na coluna 'age'
df_limpo = df_limpo.dropna(subset=['age'])

print(f"Número de valores nulos após limpeza: {df_limpo['age'].isnull().sum()}")
```

Coluna 'gender'

```python
#valores únicos da coluna 'gender'
df_limpo['gender'].unique()
```

```python
#contagem de nulos na coluna 'gender'
print(f"Número de valores nulos: {df_limpo['gender'].isnull().sum()}")
```

Coluna 'daily_social_media_time'

```python
#contagem de nulos na coluna 'daily_social_media_time'
print(f"Número de valores nulos: {df_limpo['daily_social_media_time'].isnull().sum()}")
```

```python
#substituir os nulos pela mediana
mediana = df_limpo['daily_social_media_time'].median()

df_limpo['daily_social_media_time'] = df_limpo['daily_social_media_time'].fillna(mediana)
```

```python
#contagem de nulos na coluna 'daily_social_media_time' (após limpeza)
print(f"Número de valores nulos: {df_limpo['daily_social_media_time'].isnull().sum()}")
```

```python
#tipo de valores da coluna daily_social_media_time
print(f"Tipo de dado: {df_limpo['daily_social_media_time'].dtype}")
```

```python
#verificar se os valores estão faixa de 0 a 24
df_limpo['daily_social_media_time'].describe()
```

```python
#arredondar as horas para duas casas decimais
df_limpo['daily_social_media_time'] = df_limpo['daily_social_media_time'].round(2)

df_limpo['daily_social_media_time'].head()
```
